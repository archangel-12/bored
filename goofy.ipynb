```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# Create Basemap object
map = Basemap(llcrnrlon=3.2, llcrnrlat=50.7, urcrnrlon=7.5, urcrnrlat=53.7,
              resolution='l', projection='merc')

# Draw map elements
map.drawcoastlines()
map.drawcountries()
map.fillcontinents(color='lightgray', lake_color='white')
map.drawmapboundary(fill_color='white')

# Display the map
plt.show()
```

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt

# Create Basemap object
map = Basemap(llcrnrlon=-130, llcrnrlat=22, urcrnrlon=-65, urcrnrlat=50,
              resolution='l', projection='merc')

# Draw the map with a soft white border
map.drawcoastlines(linewidth=0.5)
map.drawcountries(linewidth=0.5)
map.drawstates(linewidth=0.2)
map.drawmapboundary(fill_color='white', linewidth=0.5)

# Fill the United States with white color
map.fillcontinents(color='white', lake_color='white')

# Draw Alaska separately
map.drawcoastlines(linewidth=0.2)
map.drawstates(linewidth=0.2)

# Display the map
plt.show()
```
